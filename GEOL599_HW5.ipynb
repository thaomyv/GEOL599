```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import xarray as xr
import seaborn as sns
```

```python
df = pd.read_csv('LA_AQS_2023.csv')
```

```python
df_O3 = df[(df['Parameter Name']=='Ozone') & (df['Duration Description']=='1 HOUR')]
df_O3.head()
```

```python
sns.set_theme(style = 'whitegrid', font_scale=1.25)
oplt = sns.histplot(data=df_O3, x='Arithmetic Mean', binwidth=0.002,color='black')
oplt.set_xlabel('Ozone (ppm)')
oplt.set_ylabel('Count')
oplt.set_title('Histogram of Ozone Concentrations')
```

#### i. A histogram of the O3 concentrations
Hourly O3 concentrations in the year 2023 ranges from around 0.008 to 0.046 ppm (or 8 to 46 ppb) and the most common O3 concentrations recorded are around 0.024 to 0.034 ppm (or 24 to 34 ppb).

```python
df_NO2 = df[(df['Parameter Name']=='Nitrogen dioxide (NO2)') & (df['Duration Description']=='1 HOUR')]
df_NO2.head()
```

```python
sns.set_theme(style = 'whitegrid', font_scale=1.25)
nplt = sns.kdeplot(data=df_NO2, x='Arithmetic Mean')
nplt.set_xlabel('Nitrogen Dioxide (ppb)')
nplt.set_ylabel('Density')
nplt.set_title('Histogram of Nitrogen Dioxide Concentrations')
```

#### ii. A density plot of NO2 concentrations
The peak or mode of the density curve of hourly NO2 concentrations is at 10 ppb, which means that 10 ppb is the most frequent NO2 concentration recorded. The distribution is also positively skewed, which  means that the mean and median NO2 concentrations are higher than 10 ppb.

```python
df_1 = df_O3[['Arithmetic Mean','Date (Local)']].copy()
df_1.rename(columns={'Arithmetic Mean':'O3'}, inplace=True)
```

```python
df_1
```

```python
df_2 = df_NO2[['Arithmetic Mean','Date (Local)']].copy()
df_2.rename(columns={'Arithmetic Mean':'NO2'}, inplace=True)
```

```python
df_2
```

```python
df_all = df_1.merge(df_2, on=['Date (Local)'])
```

```python
df_all
```

```python
sns.set_theme(style = 'whitegrid', font_scale=1.25)
sp = sns.scatterplot(data=df_all, x='NO2', y='O3', color='blue')
sp.set_xlabel('Nitrogen Dioxide (ppb)')
sp.set_ylabel('Ozone (ppm)')
sp.set_title('Scatterplot of O3 vs. NO2')
```

#### iii. A scatterplot of O3 vs NO2 concentrations
There appears to be a negative relationship between O3 and NO2 concentrations, as the ozone concentrations are lower when NO2 concentrations are higher.

```python
df_1['Date (Local)'] = pd.to_datetime(df_1['Date (Local)'])
df_high = df_1[df_1['O3'] > 0.035]
```

```python
df_high['Month'] = df_high['Date (Local)'].dt.month
```

```python
df_high.drop_duplicates(subset=['Date (Local)'], inplace=True) #in case one day has multiple hourly readings over 35 ppb
```

```python
df_high
```

```python
sns.set_theme(style = 'whitegrid', font_scale=1.25)
bp = sns.countplot(data=df_high, x='Month')
bp.set_xlabel('Month')
bp.set_ylabel('Days')
bp.set_title('Number of Days in 2023 with O3 > 35 ppb')
m = ['2','3','4','5','6','7','8','9']
ml = ['Feb', 'Mar', 'April', 'May', 'June', 'July', 'Aug', 'Sep']
plt.xticks(m, ml)
```

#### iv. A barchart of the number of days per month with O3 > 35 ppb
Days with O3 concentration recording of 35 ppb or more in 2023 occurred in the months of February to September. April had the most over 35 ppb days, with a count of 11.
